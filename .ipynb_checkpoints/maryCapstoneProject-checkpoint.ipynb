# Analyzing Underemployment Rate  

```python
data = pd.read_csv('under_employment_rate.csv') #Load CSV file into memory
```

```python
data
```

```python
data = data[['group','Date','Value']] #Selecting the columns needed for the analysis
data.head()
```

```python
data_pivot= data.pivot_table(index ='Date',columns = 'group', values = 'Value') #pivot table
```

```python
data_pivot
```

```python
data1 = data_pivot.groupby('Date')  #grouping data
```

```python
data1.count()
```

```python
data1.head()
```

# Analyzing other countries unemployment rate

```python
data2 = pd.read_csv('world_unemployment_rate.csv')
```

```python
data2
```

# Tidying up data

```python
data2 = data2[['Country Name','2010','2011','2012','2013','2014','2015','2016','2017','2018']]
```

```python
data2.set_index('Country Name', inplace = True)
```

```python
data2
```

```python
data2 = data2.dropna(how = 'any') #droping missing data
```

```python
data2.isnull() #checking for missing data
```

```python
data3 = data2.loc[['China','India','United States','Canada','Nigeria']] #selecting rows needed for analysis from the dataframe
```

```python
data3.T #transpose dataframe
```

```python
data4=data3.groupby('Country Name') #grouping dataframe
```

```python
data4.describe() 
```

```python
data4 = data4.agg(np.mean)
```

```python
data4
```

```python
data4.plot(kind='bar', figsize=(15, 6), rot=90) 
plt.xlabel('Countries')
plt.ylabel('Number of Unemployment Rate')
plt.title('Comparism of unemployment rate with other developed countries from 2010 to 2018') 
plt.show()
```
