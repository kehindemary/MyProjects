```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('unemployment_rate.csv') 
```

```python
df.head()
```

```python
col_drop = ['country','indicator','group','Unit']
df.drop(col_drop, inplace = True, axis=1)
```

```python
df.head()
```

```python
Date = df['Date'].str.extract(r'^(\d{4})', expand=False)
Date = df.sort_values('Date')
```

```python
df = df['Date']=Date
```

```python
df.head()
```

```python
bins = [2010,2011,2012,2013,2014,2015,2016,2017,2018]
plt.title('Unemployement Rate')
plt.xlabel('Year')
plt.ylabel('Value')
plt.hist(df, bins, histtype = 'bar', rwidth = 0.8)
plt.show()
```
