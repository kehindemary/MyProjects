```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('world_unemployment_rate.csv')
df1
```

```python
world_unemployment_rate_df = df1[['Country Name','2010','2011','2012','2013','2014','2015','2016','2017','2018']]
world_unemployment_rate_df.head()
```

```python
world_unemployment_rate_df = pd.melt(world_unemployment_rate_df, id_vars = ['Country Name'], var_name='Year')
```

```python
world_unemployment_rate_df
```

```python
world_unemployment_rate_df = world_unemployment_rate_df.dropna(how = 'any') #droping missing data
world_unemployment_rate_df
```

```python
world_unemployment_rate_df.set_index('Country Name', inplace = True)
world_unemployment_rate_df
```

```python
world_unemployment_rate_df1 = world_unemployment_rate_df.loc[['China','India','United States','Canada','Nigeria']] #selecting rows needed for analysis from the dataframe
world_unemployment_rate_df1
```

```python
world_unemployment_rate_df1.reset_index()
```

```python
world_population_df = pd.read_csv('world_population.csv')
world_population_df.head()
```

```python
world_population_df = world_population_df[['Country Name','2010','2011','2012','2013','2014','2015']]
world_population_df.head()
```

```python
world_population_df = pd.melt(world_population_df, id_vars = ["Country Name"], var_name='Year')
world_population_df.set_index('Country Name', inplace = True)
world_population_df.head()
```

```python
world_population_df =world_population_df.loc[['China','India','United States','Canada','Nigeria']] #selecting rows needed for analysis from the dataframe
world_population_df
```

```python
world_population_df = world_population_df.reset_index()
world_population_df.head()
```

```python
new = world_unemployment_rate_df1.merge(world_population_df, on = ['Country Name','Year'])
new
```

```python
new = new.rename(columns={'Year_x':'Year1','value_x': 'No_of_unemployment','Year_y':'Year2','value_y':'Population'})
new
```

```python
new = new.pivot_table(index = 'Country Name',columns=['Year1','Year2'], values = ['No_of_unemployment','Population'])
new
```

```python
new = new.groupby(['Country Name','Year1','No_of_unemployment'])
new.count()
```
