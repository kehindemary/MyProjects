```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #importing the necessary libraries for this project
```

```python
df = pd.read_csv('unemployment_rate.csv')
df.head()
```

```python
df = df[['Date','Value']]   #selecting the columns needed for the analysis
df.head() #to returns the first 5 records in the dataset
```

```python
extr = df['Date'].str.extract(r'^(\d{4})', expand=False) #extract the date and getting rid of unwanted characters
```

```python
extr
```

```python
df['Date'] = extr #assigning the extracted date to the date column
```

```python
df['Date']
```

```python
df.head(30)
```

```python
df = df.set_index('Date') #set the index to date
```

```python
df.head()
```

```python
df = df.sort_index() #sorting the index
```

```python
df.head(50)
```

```python
gp1 = df.groupby('Date') #grouping dataframe by date
gp1
```
